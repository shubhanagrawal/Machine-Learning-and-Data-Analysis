```python
import pandas as pd
data = pd.read_csv(r"D:\CSV FILES\new_file.csv")


type_new = pd.get_dummies(data['type'], drop_first=True)
data_new = pd.concat([data, type_new], axis=1)
data_new.head()


X = data_new.drop(['isFraud', 'type', 'nameOrig', 'nameDest'], axis=1)
y = data_new['isFraud']



from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier
from sklearn.metrics import roc_auc_score as ras
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier

# Split dataset into training & testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Define models
models = [
    LogisticRegression(),
    XGBClassifier(),
    RandomForestClassifier(n_estimators=7, criterion='entropy', random_state=7)
]

# Train and evaluate models
for model in models:
    model.fit(X_train, y_train)
    print(f'{model} : ')
    
    # Check if model supports predict_proba
    if hasattr(model, "predict_proba"):
        train_preds = model.predict_proba(X_train)[:, 1]
        test_preds = model.predict_proba(X_test)[:, 1]
    else:
        train_preds = model.decision_function(X_train)
        test_preds = model.decision_function(X_test)

    print('Training Accuracy : ', ras(y_train, train_preds))
    print('Validation Accuracy : ', ras(y_test, test_preds))
    print()

```

```python
from sklearn.metrics import ConfusionMatrixDisplay
import matplotlib.pyplot as plt

cm = ConfusionMatrixDisplay.from_estimator(models[1], X_test, y_test) 

cm.plot(cmap='Blues') 

plt.show()
```
